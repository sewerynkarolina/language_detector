### TWORZĘ LISTĘ PLIKÓW ZE ŚĆIEŻKĄ- ZE WSZYSTKICH FOLDERÓW

```python
import pandas as pd
import os

#CSV-ka, która zawiera listę plików i etykiety
lista = pd.read_csv('Etykiety.csv',header=None)

#Lista etykiet - ważne, żeby były po kolei bo wrzucam je potem do data framu jako lista
list_of_countries = list(pd.Series(lista[1]))

#Listka krajów skąd mamy dane - ważne przy przechodzeniu przez foldery -> lista folderów
folder_list = list(pd.Series(lista[1]).unique())

#Ścieżki do każdego z języków
paths = [os.getcwd()+'/'+x for x in folder_list]

#Ściężka + plik.pdf
file_paths = []

#Nazwy plików po kolei - w kolejności jak języki, tak będę nazywać indeksy dataframu - więc, ważne żeby kolejność jak u języków
ind_names = []

#Dwie pętle -> wszystkie pliki z folderu:
for path in paths:
    for filename in os.listdir(path):
        file_paths.append(path+'/'+filename)
        ind_names.append(filename)
        
#Jeszcze muszę dopisać z którego folderu
file_paths
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import pdftotext
```

```python
#Element listy zawiera pełny tekst jednego z artykułów
text_list = []
for text in file_paths:
    with open(text, "rb") as f:
        pdf = pdftotext.PDF(f)
        el_of_list = ''
        #Ponieważ page in pdf  - to jest strona z artykułu to łącze stringi, pewnie to można lepiej
        for page in pdf:
            el_of_list = el_of_list+page
        text_list.append(el_of_list)
```

```python
#Sklearnem tworzę macierz
vec = CountVectorizer()
X = vec.fit_transform(text_list)
new= pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

```python
new.index = ind_names
#Jakaś głupia nazwa, żeby zapewnić, że takiej nie było w tekście
#Nazywam ostatnią kolumnę krajem
new['@C#ountry'] = list_of_countries
```
